# Conference Dataset Analysis

```python
data_path = '/home/peterfaganmath/ht09_contact_list.txt'
parsed_data_path = '/home/peterfaganmath/conference.txt'
randomised_data_path = '/home/peterfaganmath/randomconference.txt'

df = pd.read_csv(data_path ,sep='\t',header=None)
df.columns = ['time','source','target']


time_unit = 3600
max_window=59
for i in range(0,max_window):
    df.loc[(df['time']>(time_unit)*i)&(df['time']<=(time_unit)*(i+1)), 'Time'] = i+1  
df = df.drop(['time'],axis=1)
df = df.astype(int)
df = df.drop_duplicates()
Time = df['Time'].tolist()
target = df['target'].tolist()
df.to_csv(parsed_data_path, header=None, index=None, sep='\t', mode='a')
```

```python
df = df.sample(frac=1,replace=False)
df = df.reset_index()
df = df.drop(['index','Time','target'],axis=1)
df['target'] = target
df['Time'] = Time
df.to_csv(randomised_data_path, header=None, index=None, sep='\t', mode='a')
```
